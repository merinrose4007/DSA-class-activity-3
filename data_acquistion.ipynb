```python
def alert_high_temp(row):
  if(row['Temp_C']>3):
    print(f"Alert!!! Temperature is too high: {row['Temp_C']} Celcius")
    display(Audio('/content/Kiliye Kiliye - Afrobeat X Kalyani Priyadarshan (Video)  The Independeners  Aa Raathri  S Janaki.mp3', autoplay = True))
    return True
  else:
    print(f"Temperature is normal:{row['Temp_C']} Celcius")



def stream_with_alert(data, delay=1):
  for idx, row in data.iterrows():
    print(f"{row['Date_Time']} - Temperature : {row['Temp_C']}Celsius")
    if alert_high_temp(row):
      print("High temperature alert triggered. Stopping Stream processing")
      break
    time.sleep(delay)
```

```python
stream_with_alert(data)
```

Data Visualisation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
month = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
visit = [12500, 14200, 16000, 10500, 21000, 19700, 22300, 23000, 20200, 15000, 22000, 20000]
conversion_rate = [2.5, 2.8, 3.1, 3.0, 3.5, 3.2, 3.8, 4.0, 3.6, 4.0, 3.8, 3.2]

plt.plot(month, visit, '*-', color = 'yellow') # o- markers
plt.plot(month, conversion_rate, '*-', color='blue')

avg_visit = np.mean(visit)
plt.axhline(y = avg_visit, color = 'red', linestyle = '--')

max_visit = np.argmax(conversion_rate)
month_index = month[max_visit]

plt.axvline(x = month_index, color = 'violet', linestyle = '--')

plt.title('Monthely Subscribers')
plt.xlabel('Months')
plt.ylabel('subscribers')
plt.legend(['Visits', 'Conversion Rate'])
plt.grid(True, color = 'pink')
plt.show()
```

```python
max_visit = np.max(conversion_rate)
max_visit
```

```python
plt.figure(figsize=(10,4), facecolor= 'yellow')
plt.bar(month, visit, color = 'orange', edgecolor = 'black')
sales = [1500, 1300, 1200, 1700, 1800, 1900, 1100, 1600, 1300, 1500, 1600, 1900]
plt.bar(month, sales, color = 'green', edgecolor= 'blue')
plt.xticks(rotation = 45) # x axis name slanted
plt.legend(['Visit', 'Sales'])
plt.show()
```

```python
hours = [2, 3, 4, 5, 6, 8, 3, 12, 15]
mark = [20, 30, 40, 60, 100, 30, 100, 100, 1]
plt.figure(figsize=(10,4), facecolor='pink')
plt.scatter(hours, mark, color='Red')
plt.xlabel('Hours', color ='Red')
plt.ylabel('Score', color ='Red')
```

```python
step1 = np.random.randint(500, 1000, 10)
step2 = np.random.randint(1000, 5000, 30)
step3 = np.random.randint(5000, 14000, 20)
steps = np.concatenate([step1, step2, step3])

plt.hist(steps, color= 'skyblue', edgecolor='black', bins=6)
```

```python
cars = ['BMW', 'FORD', 'TESLA', 'Marcedes', 'AUDI', 'JAGUAR']
sales = [25, 17, 35, 29, 12, 41]
plt.pie(sales, labels=cars, shadow=True, autopct='%1.1f%%')
#% → indicates formatting (old-style printf formatting).
#1.1f → means floating point number with 1 digit before the decimal, 1 digit after the decimal.
#%% → prints a literal % sign (because % alone is reserved for formatting).
```
